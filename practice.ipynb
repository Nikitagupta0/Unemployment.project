```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Unemployment in India.csv",parse_dates=[' Date'])
```

```python
data.head()
```

```python
data
```

```python
print(data.isnull().sum())
```

```python
data.columns = ["States","Date","Frequency",
                "Estimated Unemployment Rate",
                "Estimated Employed",
                "Estimated Labour Participation Rate",
                "Region"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed",hue="Region",data=data)
plt.show()                
```

```python
#plt.style.use('seaborn-whitegrid')
#plt.figure(figsize=(8,8))
#sns.heatmap(data.corr())
plt.show()
```

```python
data.hist(edgecolor="black",linewidth = 1.2 , figsize=(10,10))
```

```python
d = data.Region.value_counts().head(10)                  # [:10]
plt.figure(figsize=(6,6))
x = list(d.index)
y = list(d.values)
x.reverse()
y.reverse()
plt.title("Estimated Unemployment in Regions")
plt.ylabel("Unemployment Count")
plt.xlabel("Regions")
plt.bar(x,y)
```

```python
data['States'].value_counts().head(10).plot.pie(autopct = '%.1f%%',
                                                shadow = True,
                                                figsize = (8,8),
                                                startangle = 90,
                                                wedgeprops = {'width': .4},
                                                radius = 1,
                                                pctdistance = .8,
                                                textprops = {'color':"black"})
plt.title('UNEMPLOYMENT RATE IN DIFFERENT STATES', fontsize = '11',color = 'c')                                                
```

```python
data['Region'].value_counts().head(10).plot.pie(autopct = '%.1f%%',
                                                shadow = True,
                                                figsize = (8,8),
                                                startangle = 90,
                                                wedgeprops = {'width': .4},
                                                radius = 1,
                                                pctdistance = .8,
                                                textprops = {'color':"black"})
plt.title('REGION WISE', fontsize = '11',color = 'c')
```

```python
data['Date'].value_counts().head(10).plot.pie(autopct = '%.1f%%',
                                                shadow = True,
                                                figsize = (8,8),
                                                startangle = 90,
                                                wedgeprops = {'width': .4},
                                                radius = 1,
                                                pctdistance = .8,
                                                textprops = {'color':"black"})
plt.title('UNEMPLOYMENT RATE IN MONTHS', fontsize = '11',color = 'c')
```

```python
data['Estimated Employed'].value_counts().head(10).plot.pie(autopct = '%.1f%%',
                                                shadow = True,
                                                figsize = (8,8),
                                                startangle = 90,
                                                wedgeprops = {'width': .4},
                                                radius = 1,
                                                pctdistance = .8,
                                                textprops = {'color':"black"})
plt.title('ESTIMATED EMPLOYED', fontsize = '11',color = 'c')
```

```python
plt.title("MONTHLY UNEMPLOYEMENT",color='red',fontsize=12)
data['Date'].value_counts().head(10).plot.pie(autopct = '%.1f%%',
                                                figsize = (8,8),
                                                pctdistance = .6,
                                                startangle = 180,
                                                textprops = {'color':"black"})
plt.show()
```
